```python

from urllib.parse import quote
from urllib import request
import json
import xlwt
 
amap_web_key = '2a080187e0099a2140c71b134f1c6f26'
poi_search_url = "http://restapi.amap.com/v3/place/text"
poi_boundary_url = "https://ditu.amap.com/detail/get/detail"

 
cityname = "上海"
classfiled = "剧本杀"
 

def getpois(cityname, keywords):
    i = 1
    poilist = []
    while True: 
        result = getpoi_page(cityname, keywords, i)
        print(result)
        result = json.loads(result)
        if result['count'] == '0':
            break
        hand(poilist, result)
        i = i + 1
    return poilist
 
def write_to_excel(poilist, cityname, classfield):
    book = xlwt.Workbook(encoding='utf-8', style_compression=0)
    sheet = book.add_sheet(classfield, cell_overwrite_ok=True)
 
    sheet.write(0, 0, 'id')
    sheet.write(0, 1, 'name')
    sheet.write(0, 2, 'location')
    sheet.write(0, 3, 'pname')
    sheet.write(0, 4, 'cityname')
    sheet.write(0, 5, 'adname')
    sheet.write(0, 6, 'address')
    sheet.write(0, 7, 'type')
    sheet.write(0, 8, 'tel')
    sheet.write(0, 9, 'postcode')

 
    for i in range(len(poilist)):
        sheet.write(i + 1, 0, poilist[i]['id'])
        sheet.write(i + 1, 1, poilist[i]['name'])
        sheet.write(i + 1, 2, poilist[i]['location'])
        sheet.write(i + 1, 3, poilist[i]['pname'])
        sheet.write(i + 1, 4, poilist[i]['cityname'])
        sheet.write(i + 1, 5, poilist[i]['adname'])
        sheet.write(i + 1,6, poilist[i]['address'])
        sheet.write(i + 1, 7, poilist[i]['type'])
        sheet.write(i + 1, 8, poilist[i]['tel'])
        sheet.write(i + 1, 9, poilist[i]['postcode'])
 
 

    book.save(r'' + cityname + "_" + classfield + '.xls')
 
 

def hand(poilist, result):
    pois = result['pois']
    for i in range(len(pois)):
        poilist.append(pois[i])
 
 
def getpoi_page(cityname, keywords, page):
    req_url = poi_search_url + "?key=" + amap_web_key + '&extensions=all&keywords=' + quote(
        keywords) + '&city=' + quote(cityname) + '&citylimit=true' + '&offset=25' + '&page=' + str(
        page) + '&output=json'
    data = ''
    with request.urlopen(req_url) as f:
        data = f.read()
        data = data.decode('utf-8')
    return data  
 
pois = getpois(cityname, classfiled)
 
write_to_excel(pois, cityname, classfiled)
print('写入成功')
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
import pandas as pd

for i in a:
    url = "https://shanghai.douban.com/events/week-1405?start="+str(i)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36'}
    ret = Request(url, headers=headers)
    res = urlopen(ret)
    aa = res.read().decode('utf-8')
    print(aa)
```

```python
titles = []
times = []
fees = []
owners = []

soup = BeautifulSoup(aa,'html.parser')
for item in soup.findAll("div", attrs={"class":"info"}):       
    title=item.find("span",attrs={"itemprop":"summary"})
    time = item.find("li",attrs={"class":"event-time"})
    fee =item.strong
    owner=item.find("a",attrs={"target":"db-event-owner"})   
    if title != None:
        if title != None:
            titles.append(title.text)
        else:
            owners.append(None)
        if time != None:
            times.append(time.text)
        else:
            owners.append(None)
        if fee != None: 
            fees.append(fee.text)
        else:
            owners.append(None)
        if owner != None: 
            owners.append(owner.text)
        else:
            owners.append(None)
                
#print(titles)   
#print(times)   
#print(fees)   
#print(owners)   
df1=pd.DataFrame({"event_name":titles,"fees":fees, "times":times, "owners":owners})           
df1     
```
